## 1.1基本存储原理  
    DRAM，动态随即存储。  
    原理：电源存储电荷的多少代表0和1.  
    通过控制行选和列选信号进行读写操作，其中读就是放电，写就是充电。
    DRAM的每一个bit只需要一个晶体管加一个电容。因为电容会有漏电现象，如果电荷不足，那么数据肯定会错乱，所以电容  
    必须周期性的刷新（预充电）。

## 2.2内存频率
    内存频率有3种指标，他们分别是：  
> 核心频率  
> 时钟频率  
> 有效数据传输频率
  
### 核心频率
  内存Cell阵列（Memory Cell Array，即内部电容）的刷新频率，他是内存的真实运行频率。  
### 时钟频率
  I/O Buffer（输入/输出缓存）的传输频率
### 有效数据传输频率
  数据传送的频率（等效频率）。

## 1.3内存基本参数
### 1.3.1物理bank
    传统内存系统为了保证CPU正常工作，必须一次传输完CPU在一个传输周期内所需要的数据，而CPU在一个传输周期能接受的数据容量就是CPU数据总线
    的位宽，这个位宽称之为物理BANK的位宽。例如cpu采用的是64bit位宽，假设内存颗粒是8bit，就需要用8颗内存颗粒构成一个物理bank。

### 逻辑bank
    DDR SDRAM的内部是一个存储阵列，检索原理就是表格，先指定行row，再指定列column，就可以准确找到所需要的单元格。对于表格（阵列）称为逻辑bank（LBANK）。
    一般的DDR都含有4个BANK或者8个BANK。
### tRCD
    实际工作中，L-BANK地址与相应的行地址是同时发出的，此时这个命令称之为“行有效”
    或者行激活（RowActive）。在此之后，将发送列地址寻址命令与具体的操作命令（读
    或写)，这两个命令也是同时发出的，所以一般都会以“读写命令”来表示列寻址。
    ps:列地址和读写信号是同时发出的～～～～～～～～～～～～～～～～～
    行有效到列有效之间的间隔就是tRCD。
### CL
    列地址潜伏期（CAS Latency）：指列地址被选中之后，会触发数据传输，从存储单元
    中输出到真正出现在内存芯片的IO接口之间的时间延迟。也就是CAS发出到DQS生成的间隔，这个时间叫tAC。
### tDQSS
    写入延迟，发出写入命令后，DQS与写入数据要等一段时间才会送达。
### tRP
    行预充电有效周期，SDRAM在进行完读写操作后，如果要对另一行进行寻址，就要将原来的有效行关闭，重新发送行列地址。而关闭现有工作行，准备打开新行的操作就叫预充电（Precharge）。预充电可以通过命令控制或者辅助设定让芯片每次读写操作之后自动进行预充电。上面有提到过，预充电就是一种对工作行中所有的存储体进行数据重写，并对行地址进行复位，同时释放S-AMP，对S-AMP中的数据进行回写，以准备新行的工作。通常地址线A10控制着是否进行读写后当前L-bank自动进行预充电，这个属于辅助操作，单独的充电命令中，A10控制着是对指定的L-Bank还是所有的L-Bank进行预充电，前者提供L-Bank地址，后者只需要将A10置于高电平，发出预充电命令之后，要经过一段时间才能允许RAS行有效命令打开新行，这个时间叫tRP（Precharge command Period，预充电有效时间），在datasheet中也是有表明这个时间周期的。
### BL
    突发（Brust）是指在同一行中相邻的单元连续进行数据传输的方式，连续传输的周期数就是突发长度。
### DQM
    数据掩码，屏蔽不需要的数据，例如32bit由4个DQM线控制。
### DQS
    数据选取脉冲，用来在一个时钟周期内准确的区分出每个传输周期，并便于接收方准确接收数据。每一颗芯片都有一个DQS信号线，它是杀u嗯想的，在写入时它用来传送由北桥发来的DQS信号，读取时，则由芯片生成DQS向北桥发送。
### DLL
    延迟锁定回路。DLL用于生成一个延迟量给内部时钟。最终目的是使内外时钟同步。
